```python
import pandas as pd
import numpy as np
```

**Data Loading**

```python
# BRCA

brca = pd.read_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/brca_sig.csv', sep=',')
brca = brca[["cancer_type", "ncbi_gene_id", "gene_symbol", "fdr_adjusted_p_value", "log2_fold_change"]]
brca
```

```python
# COAD

coad = pd.read_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/coad_sig.csv', sep=',')
coad = coad[["cancer_type", "ncbi_gene_id", "gene_symbol", "fdr_adjusted_p_value", "log2_fold_change"]]
coad
```

```python
# KIRC

kirc = pd.read_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/kirc_sig.csv', sep=',')
kirc = kirc[["cancer_type", "ncbi_gene_id", "gene_symbol", "fdr_adjusted_p_value", "log2_fold_change"]]
kirc
```

```python
# LUAD

luad = pd.read_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/luad_sig.csv', sep=',')
luad = luad[["cancer_type", "ncbi_gene_id", "gene_symbol", "fdr_adjusted_p_value", "log2_fold_change"]]
luad
```

**Identify Shared DEGs Across Cancer Types**

```python
shared = pd.merge(pd.merge(pd.merge(brca, coad, on = ['ncbi_gene_id', 'gene_symbol']), kirc, on = ['ncbi_gene_id', 'gene_symbol']), luad, on = ['ncbi_gene_id', 'gene_symbol'])
shared.head()
```

```python
list(shared.columns)
```

```python
sharedDF = shared.set_axis(['brca', 'ncbi_gene_id', 'gene_symbol', 'brca_fdr', 'brca_lfc', 'coad', 'coad_fdr', 'coad_lfc', 'kirc', 'kirc_fdr', 'kirc_lfc', 'luad', 'luad_fdr', 'luad_lfc'], axis = 1, inplace = False)
subsetDF = sharedDF[['ncbi_gene_id', 'gene_symbol', 'brca_fdr', 'brca_lfc', 'coad_fdr', 'coad_lfc', 'kirc_fdr', 'kirc_lfc', 'luad_fdr', 'luad_lfc']]
subsetDF.index = np.arange(1, len(subsetDF) + 1)
subsetDF
```

**Export Shared DEGs**

```python
subsetDF.to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/shared_DEGs.csv', index = False)
```

**Export NCBI Gene IDs & Gene Symbols of Shared DEGs**

```python
subsetDF['ncbi_gene_id'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/shared_NCBI_gene_IDs.csv', index = False)
subsetDF['gene_symbol'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/shared_gene_symbols.csv', index = False)
```

**Identify Cancer-Specific DEGs**

```python
subsetDF.reset_index(drop=True, inplace=True)
sharedID = subsetDF['ncbi_gene_id'].to_frame()
sharedID['ncbi_gene_id'].tolist()
sharedID.head()
```

```python
# BRCA

brcaDiff = brca[~brca['ncbi_gene_id'].isin(sharedID['ncbi_gene_id'])]
brcaDiff
```

```python
# COAD

coadDiff = coad[~coad['ncbi_gene_id'].isin(sharedID['ncbi_gene_id'])]
coadDiff
```

```python
# KIRC

kircDiff = kirc[~kirc['ncbi_gene_id'].isin(sharedID['ncbi_gene_id'])]
kircDiff
```

```python
# LUAD

luadDiff = luad[~luad['ncbi_gene_id'].isin(sharedID['ncbi_gene_id'])]
luadDiff
```

**Export Cancer-Specific DEGs**

```python
brcaDiff.to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/brca_diff.csv', index = False)
coadDiff.to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/coad_diff.csv', index = False)
kircDiff.to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/kirc_diff.csv', index = False)
luadDiff.to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/luad_diff.csv', index = False)
```

**Export NCBI Gene IDs & Gene Symbols of Cancer-Specific DEGs**

```python
# BRCA
brcaDiff['ncbi_gene_id'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/brcaDiff_NCBI_gene_IDs.csv', index = False)
brcaDiff['gene_symbol'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/brcaDiff_gene_symbols.csv', index = False)

# COAD
coadDiff['ncbi_gene_id'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/coadDiff_NCBI_gene_IDs.csv', index = False)
coadDiff['gene_symbol'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/coadDiff_gene_symbols.csv', index = False)

# KIRC
kircDiff['ncbi_gene_id'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/kircDiff_NCBI_gene_IDs.csv', index = False)
kircDiff['gene_symbol'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/kircDiff_gene_symbols.csv', index = False)

# LUAD
luadDiff['ncbi_gene_id'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/luadDiff_NCBI_gene_IDs.csv', index = False)
luadDiff['gene_symbol'].to_csv('/content/drive/My Drive/Colab Notebooks/BINF 3350: Genomics & Bioinformatics/luadDiff_gene_symbols.csv', index = False)
```
